```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
print("Gensim working ")
```

```python
corpus = [
    "Natural language processing is interesting",
    "Natural language processing is powerful",
    "Machine learning is related to artificial intelligence",
    "Artificial intelligence and machine learning are connected",
    "Language models predict next words"
]
```

```python
tokenized_corpus = [word_tokenize(sentence.lower()) for sentence in corpus]
tokenized_corpus
```

```python
model = Word2Vec(
    sentences=tokenized_corpus,
    vector_size=50,
    window=2,
    min_count=1,
    workers=4
)
```

```python
model.wv["language"]
```

```python
model.wv.most_similar("machine")
```

```python
model.wv.similarity("machine", "learning")
```
