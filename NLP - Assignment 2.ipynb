```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
```

```python
corpus = [
    "Natural language processing is very interesting.",
    "The model predicts legal grievance categories.",
    "She studies computer science and artificial intelligence."
]
```

```python
tokenized_data = [word_tokenize(sentence) for sentence in corpus]
tokenized_data
```

```python
pos_tagged_data = [pos_tag(tokens) for tokens in tokenized_data]
pos_tagged_data
```

```python
for i in range(len(corpus)):
    print("Sentence:", corpus[i])
    print("POS Tags:", pos_tagged_data[i])
    print()
```
