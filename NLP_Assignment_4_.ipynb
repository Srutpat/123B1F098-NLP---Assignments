```python
import nltk
from nltk.tokenize import word_tokenize
from gensim.models import Word2Vec
import numpy as np
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
corpus = [
    "I love coding in python",
    "Python is great for machine learning",
    "Python is used for data science",
    "Coding in python is fun",
    "Machine learning with python is powerful",
    "I enjoy python programming",
    "Python helps in data analysis",
    "Learning python improves coding skills",
    "Python is popular in software development",
    "Coding and programming go hand in hand"
]
```

```python
stop_words = {'is', 'in', 'for', 'and', 'the', 'i'}

processed_corpus = []
for sentence in corpus:
    words = sentence.lower().split()
    words = [w for w in words if w not in stop_words]
    processed_corpus.append(words)
```

```python
model = Word2Vec(
    sentences=processed_corpus,
    vector_size=100,
    window=3,
    min_count=1,
    sg=1,
    epochs=200
)
```

```python
print("\nVector for 'python':\n", model.wv['python'])
```

```python
def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2))

word1 = "python"
word2 = "coding"

vec1 = model.wv[word1]
vec2 = model.wv[word2]

similarity = cosine_similarity(vec1, vec2)

print(f"\nSimilarity between '{word1}' and '{word2}': {similarity}")
```

```python
print("\nWords similar to 'python':")
print(model.wv.most_similar("python"))
```
