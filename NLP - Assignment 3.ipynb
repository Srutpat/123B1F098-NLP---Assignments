```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import word_tokenize
from nltk.util import ngrams
from nltk.probability import FreqDist
```

```python
text = "natural language processing is fun and natural language models are powerful"
tokens = word_tokenize(text.lower())
tokens
```

```python
bigrams = list(ngrams(tokens, 2))
bigrams
```

```python
freq_dist = FreqDist(bigrams)
freq_dist
```

```python
def predict_next_word(word):
    candidates = [pair[1] for pair in freq_dist if pair[0] == word]
    if candidates:
        return max(candidates, key=lambda w: freq_dist[(word, w)])
    else:
        return None

predict_next_word("natural")
```
