```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, GlobalAveragePooling1D, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# 1. Sample dataset
texts = [
    "win money now",
    "claim your prize",
    "hello how are you",
    "let us meet tomorrow",
    "free cash offer",
    "are you coming today"
]
labels = [1, 1, 0, 0, 1, 0]  # 1 = spam, 0 = not spam
```

```python
# 2. Tokenization
tokenizer = Tokenizer()
tokenizer.fit_on_texts(texts)
word_index = tokenizer.word_index

sequences = tokenizer.texts_to_sequences(texts)
max_len = 5
X = pad_sequences(sequences, maxlen=max_len)

y = np.array(labels)
```

```python
# 3. Model with Word Embeddings
model = Sequential()
model.add(Embedding(input_dim=len(word_index)+1, output_dim=8, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# 4. Train
model.fit(X, y, epochs=50, verbose=1)
```

```python
# 5. Test (Real-time prediction)
def predict_message(msg):
    seq = tokenizer.texts_to_sequences([msg])
    padded = pad_sequences(seq, maxlen=max_len)
    pred = model.predict(padded)[0][0]

    if pred > 0.5:
        return "Spam"
    else:
        return "Not Spam"

# Example
print(predict_message("free money now"))
print(predict_message("let's have lunch"))
```
