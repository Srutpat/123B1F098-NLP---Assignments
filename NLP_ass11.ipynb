```python
import nltk
from nltk import word_tokenize, pos_tag, RegexpParser

# Download required resources (run once)
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')   # required in some newer environments
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')

```

```python
# Paragraph
text = """The rapid growth of artificial intelligence is transforming many industries.
Modern machine learning models can analyze large amounts of data efficiently.
The skilled data scientists build innovative solutions for complex problems.
Many global companies invest heavily in advanced technologies.
The increasing demand for automation is changing the job market significantly.
These technological advancements create new opportunities for businesses and individuals."""
```

```python
# Tokenize
tokens = word_tokenize(text)

# POS tagging
tagged = pos_tag(tokens)

# Define grammar
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"

# Chunking
chunk_parser = RegexpParser(grammar)
tree = chunk_parser.parse(tagged)
```

```python
# Extract noun phrases
noun_phrases = []
for subtree in tree.subtrees(filter=lambda t: t.label() == 'NP'):
    np = " ".join(word for word, tag in subtree.leaves())
    noun_phrases.append(np)

print("Noun Phrases:")
for np in noun_phrases:
    print(np)
```
