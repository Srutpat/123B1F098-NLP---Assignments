```python
import nltk
from nltk import word_tokenize, pos_tag
from nltk.chunk import RegexpParser
```

```python
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('punkt_tab')
```

```python
paragraph = """Natural language processing techniques are widely used in artificial intelligence applications.
The intelligent system extracts meaningful information from large datasets."""
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
chunk_parser = RegexpParser(grammar)

noun_phrases = []
```

```python
for sentence in sentences:

    words = word_tokenize(sentence)

    pos_tags = pos_tag(words)

    tree = chunk_parser.parse(pos_tags)

    for subtree in tree.subtrees():
        if subtree.label() == 'NP':
            np = " ".join(word for word, tag in subtree)
            noun_phrases.append(np)
```

```python
print("Extracted Noun Phrases:\n")
for np in noun_phrases:
    print("-", np)
```
